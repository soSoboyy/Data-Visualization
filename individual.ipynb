# Feature Engineering
#### Using OHE : 
- The dataset has some column that are categorical rather than numerical.
- We don't have categorial values that shows any kind or 'order', so AutoEncoder() is not the best option.
- Since we need to create our ML algorithms, we don't want to confuse the algorithms so OHE is the best option.

```python
# Display basic info
print(cyberAttacks.info())
```

```python
# Classifying financial loss by severity
def loss_level(value):
    if value < 10:
        return "Low"
    elif value <= 30:
        return "Medium"
    else:
        return "High"

# Apply the function to create a new column
cyberAttacks["Loss Severity"] = cyberAttacks["Financial Loss (in Million $)"].apply(loss_level)

# Preview the changes
cyberAttacks.head(5)
```

```python
# Hot encoding categorical features
encodedCyberAttacks =  pd.get_dummies(cyberAttacks, columns=['Country','Attack Type','Target Industry','Attack Source','Security Vulnerability Type','Defense Mechanism Used'],drop_first=False)

encodedCyberAttacks.info()
```

```python
#Dropping all NAN values
encodedCyberAttacks.dropna(inplace=True)
```

```python
# calculating the correlation matrix
correlation_matrix = encodedCyberAttacks.corr()
```

```python
# calculating the correlation matrix

plt.figure(figsize=(25, 22))
sns.heatmap(correlation_matrix, annot=True, cmap='Spectral', fmt='.2f',linewidths=0.5)
plt.title('Correlation Heatmap of Encoded features', fontsize=14)
plt.show()
```
